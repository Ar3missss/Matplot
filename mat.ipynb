```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

# Resize your Graph 
plt.figure(figsize=(5,3),dpi=100)

plt.plot(x,y, 
         label='2x',
         color='Red',
         linestyle='--',
         linewidth=2,
         marker='*',
         markersize='12',
         markeredgecolor='Blue')

## line number 2

x2=np.arange(0,4,0.5) # np.arrange(start, stop, step) this will givw tota 8 values

plt.plot(x2[:6],x2[:6]**2, # its called slicing it will give u value 0 to 5 index
         color='Red',
         label='x2')

plt.plot(x2[5:],x2[5:]**2, # it will give u value after index 5
         color='Red',
         label='x2',
         linestyle='--')



plt.title('Our First Graph',
          fontdict={'fontname':'Comic Sans MS',
                    'fontsize':20,
                    'color':'Green'})

plt.xlabel('X-Axis',
           color='Green')


plt.ylabel('Y-Axis',
           color = 'purple' )

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.savefig('outputs/mygraph.png',dpi=300) # recommend dpi when u export ur graph is 300 so ur graph not look pixilated
plt.legend()
plt.show()
```

 BARCHART

```python
x = ['A','B','C']
y = [1,4,2]

bars=plt.bar(x,y)

patterns = ['/','O','*']
for bar ,pattern in zip(bars,patterns):
        bar.set_hatch(pattern)

# Things You do extra with barchart
# Another way to plot values
#bars[0].set_hatch('/')
#bars[1].set_hatch('*')
#bars[2].set_hatch('o')


plt.savefig('outputs/bar.png',dpi = 300)
plt.show()
```
