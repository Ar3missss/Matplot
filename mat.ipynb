```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=[0,1,2,3,4]
y=[0,2,4,6,8]

plt.plot(x,y, 
         label='2x',
         color='Red',
         linestyle='--',
         linewidth=2,
         marker='*',
         markersize='12',
         markeredgecolor='Blue')

plt.title('Our First Graph',
          fontdict={'fontname':'Comic Sans MS',
                    'fontsize':20,
                    'color':'Green'})

plt.xlabel('X-Axis',
           color='Black')


plt.ylabel('Y-Axis')

plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8,10])

plt.legend()
plt.show()
```
