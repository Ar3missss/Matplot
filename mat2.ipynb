```python
import matplotlib.pyplot as plt
import pandas as pd 

gas = pd.read_csv('data/gas_prices.csv')
petrol =gas.copy()

plt.title('Gas Prices Over Time (in USD)')

# Without Loop
plt.plot(gas.Year,gas.USA , 'g.-' , label ='USA') # here 'b.-' is color,marker,line
plt.plot(gas.Year,gas.UK , 'b.-', label ='UK')
plt.plot(gas.Year,gas.France ,'r.-' , label ='France')
plt.plot(gas.Year ,gas.Germany , 'y.-', label ='Germany')
plt.plot(gas.Year ,gas.Japan , '.-', label ='Japan')

plt.xticks(gas.Year[::3].tolist()+[2011])

plt.xlabel('Year')
plt.ylabel('US Dollars')

plt.savefig('outputs/gas_prices.png',dpi =300)

plt.legend()
plt.show()
```
